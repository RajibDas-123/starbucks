# Data Sets

The data is contained in three files:

* portfolio.json - containing offer ids and meta data about each offer (duration, type, etc.)
* profile.json - demographic data for each customer
* transcript.json - records for transactions, offers received, offers viewed, and offers completed

Here is the schema and explanation of each variable in the files:

**portfolio.json**
* id (string) - offer id
* offer_type (string) - type of offer ie BOGO, discount, informational
* difficulty (int) - minimum required spend to complete an offer
* reward (int) - reward given for completing an offer
* duration (int) - time for offer to be open, in days
* channels (list of strings)

**profile.json**
* age (int) - age of the customer 
* became_member_on (int) - date when customer created an app account
* gender (str) - gender of the customer (note some entries contain 'O' for other rather than M or F)
* id (str) - customer id
* income (float) - customer's income

**transcript.json**
* event (str) - record description (ie transaction, offer received, offer viewed, etc.)
* person (str) - customer id
* time (int) - time in hours since start of test. The data begins at time t=0
* value - (dict of strings) - either an offer id or transaction amount depending on the record

```python
def getValue(event, value):    
    items = []
    for key, value in value.items():
        items.append(value)
    items.append(0)
    return items[:2]
```

```python
def checkIfInfluenced(ser):
    global countr
    countr += 1
    print(countr)
    user_id = ser["person"]
    time = int(ser["time"])
    offer_id = ser["offer_id"]
    duration = ser["duration"]
    
#     print(user_id, time, offer_id, duration)


    
    user_data = offer_rec_view_cleaned[offer_rec_view_cleaned.person == user_id] 
#     print(len(user_data), "Records found")
    
    offer_rec_df = user_data[(user_data.offer_id == offer_id)&(user_data.event == 'offer received') & (user_data.time <= time)]
#     print(len(offer_rec_df), "Records found Received")
#     print(offer_rec_df)
#     if(len(offer_rec_df) == 0):
#         return False
    offer_rec_time = int(offer_rec_df.iloc[-1, ]['time'])
    
#     print("Recieved time", offer_rec_time)
    
    offer_view_df = user_data[(user_data.offer_id == offer_id)&(user_data.event == 'offer viewed')
                             & (user_data.time >= offer_rec_time)&(user_data.time <= time)]
#     print(len(offer_rec_df), "Records found Viewed")
#     print(offer_view_df)
    if(len(offer_view_df) == 0):
        return False
    
    return (24 * int(duration.days) - (time - offer_rec_time)) >= 0
    
```

```python
def getOfferReceivedTime(person, time, offer_id):
    df = offer_rec_view_cleaned[(offer_rec_view_cleaned.time <= int(time))
                                &(offer_rec_view_cleaned.person == person)
                                &(offer_rec_view_cleaned.offer_id == offer_id)
                                &((offer_rec_view_cleaned.event == 'offer received'))
                                ]
    if len(df) == 0:
        return 
    return df.time.values.max()
    
```
